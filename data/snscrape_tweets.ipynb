## Importing libraries here

```python
#Importing necessary library
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
```

```python
import os
path = os.getcwd()

d = ['2020-02-01','2020-03-01','2020-04-01','2020-05-01','2020-06-01','2020-07-01','2020-08-01','2020-09-01','2020-10-01','2020-11-01','2020-12-01','2021-01-01','2021-02-01']
city = ['New York', 'Los Angeles', 'Las Vegas', 'Miami']
for j in city:
    tmp_path = path+"\\"+j+"\\"
    if not os.path.exists(tmp_path):
        os.mkdir(tmp_path)
    for i in range(12):
        script = '#wfh since:'+d[i]+' until:'+d[i+1]+' near:"'+j+'" within:70km lang:en'
        print(script)
        df = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(script).get_items(), 1000))
        df.to_csv(tmp_path+j+"_wfh_tweets_"+d[i][:7]+".csv")
        print(j,i,'finished')
        #print(x)
    
```

```python
import os
path = 'f:\\Desktop'

target_lst = []

for c in city:
    
    tmp_path = path+"\\"+c+"\\"
    os.chdir(tmp_path)
    print(os.getcwd())
    
    tmp_lst = []
    
    for i in os.listdir():
        #print(i)
        df = pd.read_csv(i)
        
        if (not df.empty):
            tmp_lst.append(df[['date', 'content','renderedContent','hashtags','likeCount']])
        
    tmp_df = pd.concat(tmp_lst, ignore_index=True, axis=0)
    tmp_df['Location'] = c
    target_lst.append(tmp_df)
    
target = pd.concat(target_lst, ignore_index=True, axis=0)
```

```python
d2 = ['2020-02-01','2020-03-01','2020-04-01','2020-05-01','2020-06-01','2020-07-01','2020-08-01','2020-09-01','2020-10-01','2020-11-01','2020-12-01','2021-01-01','2021-02-01','2021-03-01','2021-04-01','2021-05-01','2021-06-01','2021-07-01','2021-08-01','2021-09-01','2021-10-01','2021-11-01','2021-12-01','2022-01-01','2022-02-01','2022-03-01','2022-04-01']
city2 = ['New York', 'Los Angeles', 'Las Vegas', 'Miami', "Chicago",'Philadelphia']
```

```python
target_lst2 = []
path = 'f:\\Desktop'
for j in city2:
    
    tmp_path = path+"\\"+j+"\\"
    if not os.path.exists(tmp_path):
        os.mkdir(tmp_path)

    city_lst = []
    for i in range(len(d2)-1):
        script = '#wfh since:'+d2[i]+' until:'+d2[i+1]+' near:"'+j+'" within:70km lang:en'
        print(script)
        df = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(script).get_items(), 1000))
        df.to_csv(tmp_path+j+"_wfh_tweets_"+d2[i][:7]+".csv")
        
        if (not df.empty):
            city_lst.append(df[['date', 'content','renderedContent','hashtags','likeCount']])
            
    city = pd.concat(city_lst, ignore_index=True, axis=0)
    city['location'] = j
    target_lst2.append(city)

target = pd.concat(target_lst2, ignore_index=True, axis=0)
```
